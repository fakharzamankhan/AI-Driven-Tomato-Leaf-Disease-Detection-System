```python
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import json
from sklearn.metrics import confusion_matrix, classification_report
import os

# Unzip your dataset from Google Drive to local Colab environment
!unzip -q /content/drive/MyDrive/plant_dataset.zip -d ./plant_dataset

# Verify directory contents to confirm unzip worked
print("Contents of plant_dataset folder:", os.listdir('./plant_dataset'))
# Updated paths due to nested folder structure after unzip
print("Contents of train folder:", os.listdir('./plant_dataset/plant_dataset/train'))
print("Contents of valid folder:", os.listdir('./plant_dataset/plant_dataset/valid'))

# Set dataset directories (updated paths)
train_dir = './plant_dataset/plant_dataset/train'
valid_dir = './plant_dataset/plant_dataset/valid'



# Load datasets
train_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    image_size=(128, 128),
    batch_size=32,
    label_mode="categorical",
    shuffle=True,
    seed=42
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    valid_dir,
    image_size=(128, 128),
    batch_size=32,
    label_mode="categorical",
    shuffle=False
)

# Class names (ORDER IS CRITICAL)
class_names = train_ds.class_names
print("Classes:", class_names)

# Normalize images
normalization_layer = tf.keras.layers.Rescaling(1./255)

train_ds = train_ds.map(lambda x, y: (normalization_layer(x), y))
val_ds = val_ds.map(lambda x, y: (normalization_layer(x), y))

# CNN model (balanced for small dataset)
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, 3, activation='relu', padding='same', input_shape=(128,128,3)),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Conv2D(64, 3, activation='relu', padding='same'),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Conv2D(128, 3, activation='relu', padding='same'),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dropout(0.4),

    tf.keras.layers.Dense(len(class_names), activation='softmax')
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# Early stopping
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True
)

# Train
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=30,
    callbacks=[early_stop]
)

# Save model
model.save("trained_plant_disease_model.keras")
print("Model saved successfully")

# Save history
with open("training_history.json", "w") as f:
    json.dump(history.history, f)

# Evaluation
val_loss, val_acc = model.evaluate(val_ds)
print(f"Validation Accuracy: {val_acc:.4f}")

# Confusion matrix
y_true = []
y_pred = []

for images, labels in val_ds:
    preds = model.predict(images)
    y_true.extend(tf.argmax(labels, axis=1))
    y_pred.extend(tf.argmax(preds, axis=1))

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8,8))
sns.heatmap(cm, annot=True, fmt='d',
            xticklabels=class_names,
            yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

print("\nClassification Report\n")
print(classification_report(y_true, y_pred, target_names=class_names))
```
